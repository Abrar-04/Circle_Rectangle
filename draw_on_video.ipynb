```python
## draw using co ordinate

import cv2

cap=cv2.VideoCapture(0)

width=int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height=int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

x=width//2
y=height//2

w=width//4
h=height//4

while True:
    ret,frame=cap.read()
    cv2.rectangle(frame,(x,y),(x+w,y+w),color=(25,65,255),thickness=5)
    cv2.imshow('frame',frame)
    
    if cv2.waitKey(1) & 0xff==27:
        break

cap.release()
cv2.destroyAllWindows()

```

```python
#interqcative rectangle

import cv2


def draw_rec(event,x,y,flags,param):
    
    global pt1,pt2,topLeftClick,botRightClick
    
    if event==cv2.EVENT_LBUTTONDOWN:
        
        #to reset
        if topLeftClick and botRightClick:
            pt1=(0,0)
            pt2=(0,0)
            topLeftClick=False
            botRightClick=False 
            
        #left click
        if topLeftClick==False:
            pt1=(x,y)
            topLeftClick=True
        
        #right click
        elif botRightClick==False:
            pt2=(x,y)
            botRightClick=True


#global var
pt1=(0,0)
pt2=(0,0)
topLeftClick=False
botRightClick=False                                                           


cap=cv2.VideoCapture(0)
cv2.namedWindow("TEST")
cv2.setMouseCallback("TEST",draw_rec)



while True:
    ret,frame=cap.read()
    
    #draw the circle marker
    if topLeftClick:
        cv2.circle(frame,center=pt1,radius=5,color=(0,255,0),thickness=-1)
    
    #draw rectangle
    if topLeftClick and botRightClick:
        cv2.rectangle(frame,pt1,pt2,(0,255,0),3)
        
    
    cv2.imshow('TEST',frame)
    
    if cv2.waitKey(1) & 0xff==27:
        break

cap.release()
cv2.destroyAllWindows()

```

```python
#interactive circle and rectangle

import cv2


def draw_rec_circle(event,x,y,flags,param):
    
    global pt1,pt2,topLeftClick,botRightClick,centers,click
    
    if event==cv2.EVENT_LBUTTONDOWN:
        
        #to reset
        if topLeftClick and botRightClick:
            pt1=(0,0)
            pt2=(0,0)
            topLeftClick=False
            botRightClick=False 
            
        #left click
        if topLeftClick==False:
            pt1=(x,y)
            topLeftClick=True
        
        #right click
        elif botRightClick==False:
            pt2=(x,y)
            botRightClick=True
            
    
    if event==cv2.EVENT_RBUTTONDOWN:
        centers=(x,y)
        click=False
    
    if event==cv2.EVENT_RBUTTONUP:
        click=True


#global var
pt1=(0,0)
pt2=(0,0)
topLeftClick=False
botRightClick=False
centers=(0,0)
click=False


cap=cv2.VideoCapture(0)
cv2.namedWindow("TEST")
cv2.setMouseCallback("TEST",draw_rec_circle)



while True:
    ret,frame=cap.read()
    
    #draw the circle marker
    if topLeftClick:
        cv2.circle(frame,center=pt1,radius=5,color=(0,255,0),thickness=-1)
    
    #draw rectangle
    if topLeftClick and botRightClick:
        cv2.rectangle(frame,pt1,pt2,(0,255,0),3)
        
    #center marker
    if click:
        cv2.circle(frame,center=centers,radius=50,color=(255,0,0),thickness=5)
        
    
    cv2.imshow('TEST',frame)
    
    if cv2.waitKey(1) & 0xff==27:
        break

cap.release()
cv2.destroyAllWindows()
```
